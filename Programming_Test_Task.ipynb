## Data cleaning

```python
#All columns are either integers, floats, or objects ie. strings
df.dtypes
```

```python
#Drop insignificant records with significant information missing
df = df.dropna(subset=['zipcode', 'datePostedString'])
```

```python
#Time on market is very important so we take the average of all time and input number
df = df.fillna(df.mean())
```

```python
#Fill in empty values with no description. No need to delete records in this case
df['description'] = df['description'].fillna('no description')
```

```python
#Check for any missing values again
df.isnull().sum()
```

```python
#Delete first column since there is no use for it
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.head()
```

```python
df.to_csv('RealEstate_California_Cleaned.csv', index = False)
```

## Setup and Populate Database

```python
try:
    conn = msql.connect(host='localhost', database='realestate', user='root', password='')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        print('Creating table....')
        cursor.execute("CREATE TABLE geographical_information(id varchar(255),address varchar(255),city varchar(255),state varchar(255),zipcode varchar(255))")
        print("Table is created....")
        #loop through data frame
        for i,row in geographical_df.iterrows():
            sql = "INSERT INTO realestate.geographical_information VALUES (%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")

            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```

## Validating Database Connection

```python
conn = msql.connect(host='localhost', database='realestate', user='root', password='')

def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")

select_all = "SELECT * FROM geographical_information"
test = execute_read_query(conn, select_all)

print('VALIDATING CONNECTION...')
for i in test:
    print(i)
```
